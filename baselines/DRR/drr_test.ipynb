```python
def gen_paragraph_data(datas):
    l = []
    for data in datas:
        text_id = data["id"]
        for para in data["paragraph_logicRelation"]:
            Arg1_paragraphIdx = para["Arg1"]["paragraphIdx"]
            Arg2_paragraphIdx = para["Arg2"]["paragraphIdx"]
            Connective = para["Connective"]
            Relation = para["Relation"]
            RelationType = para["RelationType"]
            Arg1_RawText = para["Arg1"]["RawText"]
            Arg2_RawText = para["Arg2"]["RawText"]
            d = {
                "ID": text_id,
                "paragraph1":{
                    "Idx":Arg1_paragraphIdx,
                    "Text":Arg1_RawText
                },
                "paragraph2":{
                    "Idx":Arg2_paragraphIdx,
                    "Text":Arg2_RawText
                },
                "Connective":Connective,
                "RelationType":RelationType,
                "Relation":Relation
                
            }
            l.append(d)
    return l
```

```python
train = gen_paragraph_data(datas[:401])
test = gen_paragraph_data(datas[-100:])
```

```python
def gen_sentence_data(datas):
    l = []
    for data in datas:
        text_id = data["id"]
        for sent in data["sentence_logicRelation"]:
            Arg1_paragraphIdx = sent["Arg1"]["paragraphIdx"]
            Arg2_paragraphIdx = sent["Arg2"]["paragraphIdx"]
            Arg1_sentenceIdx = sent["Arg1"]["sentenceIdx"]
            Arg2_sentenceIdx = sent["Arg2"]["sentenceIdx"]
            Connective = sent["Connective"]
            Relation = sent["Relation"]
            RelationType = sent["RelationType"]
            Arg1_RawText = sent["Arg1"]["RawText"]
            Arg2_RawText = sent["Arg2"]["RawText"]
            d = {
                "ID": text_id,
                "sentence1":{
                    "Idx":Arg1_paragraphIdx,
                    "sentenceIdx":Arg1_sentenceIdx,
                    "Text":Arg1_RawText
                },
                "sentence2":{
                    "Idx":Arg2_paragraphIdx,
                    "sentenceIdx":Arg2_sentenceIdx,
                    "Text":Arg2_RawText
                },
                "Connective":Connective,
                "RelationType":RelationType,
                "Relation":Relation
                
            }
            l.append(d)
    return l
```

```python
train_s = gen_sentence_data(datas[:401])
test_s = gen_sentence_data(datas[-100:])
```
